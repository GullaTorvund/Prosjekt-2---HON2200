```python
from aitextgen import aitextgen
```

```python
ai = aitextgen(model_name="gpt-small")
```

```python
import os

directory = os.fsencode("../input/poetry/")
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    print(filename)
    ai.train(f"../input/poetry/{filename}",
            line_by_line=False,
            from_cache=False,
            num_steps=10,
            learing_rate=1e-10,
            fp16=True,
            batch_size=3
            )
```

```python
ai.generate(prompt="I miss you", max_length=200)
ai.generate(prompt="All the small things", max_length=200)
```
